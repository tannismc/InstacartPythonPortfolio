# 03 Import customers dataset

```python
# Import customers.pkl
df_cust = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared data', 'customers.pkl'))
df_cust.head()
```

# 04 Merge datasets

```python
# Merge dataframes based on user_id
df_all = df_ords_prods.merge(df_cust, on='user_id', indicator = True)
df_all.head()
```

# 05 Export data

```python
# Remove _merge column from data
df_all = df_all.drop(columns=['_merge'])
```

```python
# Export df_all to pkl
df_all.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_all.pkl'))
```
