# 03 Create a loyalty flag for existing customers

```python
# Create a max_order column to determine the number of orders for each customer
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the results of the above function
df_ords_prods_merge.head(15)
```

```python
# Set the criteria and execute the code to set the flag for Loyal Customers
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag']= 'Loyal Customer'
```

```python
# Set the criteria and execute the code to set the flag for Regular Customers
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order']  <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# Set the criteria and execute the code to set the flag for New Customers
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Print the frequency of new "loyalty_flag" column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

# 04 Create a spending flag

```python
# Create a mean_prices column to determine the average price across all orders for a customer
df_ords_prods_merge['mean_prices'] = round(df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean), 2)
```

```python
# Check the results of the above function
df_ords_prods_merge.head(15)
```

```python
# Set the criteria and execute the code to flag Low spenders
df_ords_prods_merge.loc[df_ords_prods_merge['mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Set the criteria and execute the code to flag High spenders
df_ords_prods_merge.loc[df_ords_prods_merge['mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Print the frequency of new "spending_flag" column
df_ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

# 05 Create an order frequency flag

```python
# Create a median_days_since_order column to determine the average price across all orders for a customer
df_ords_prods_merge['median_days_since_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the results of the above function
df_ords_prods_merge.head(15)
```

```python
# Set the criteria and execute the code to set the Non-frequent customer flag
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_order'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
# Set the criteria and execute the code to set the Regular customer flag
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_order'] > 10) & (df_ords_prods_merge['median_days_since_order'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
# Set the criteria and execute the code to set the Frequent customer flag
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_order'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Check the results of the above function
df_ords_prods_merge.head(15)
```

```python
# Print the frequency of new "order_frequency" column
df_ords_prods_merge['order_frequency'].value_counts(dropna=False)
```
